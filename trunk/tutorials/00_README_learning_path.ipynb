## What is PhlyGreen?

PhlyGreen sizes each subsystem of a (possibly hybrid-electric / hydrogen) aircraft so the vehicle can fly a given mission within its design constraints. Because component masses depend on mission performance, which depends on take-off weight, the sizing is an **iterative weight-convergence loop** (Brent's method). It is developed at Sapienza University of Rome.

The public API we use throughout:

- `pg.build_aircraft()` — wire up a fresh aircraft.
- `aircraft.configure(config)` — run the full sizing loop.
- `aircraft.results()` — a structured `AircraftResults` (masses, powers, energy).
- `pg.run_design(config)` / `pg.evaluate(base, apply, x)` — stateless one-shot design (safe to call in a loop). Baseline configs live in `examples/common.py`.

Useful background reading: any aircraft-design text (Raymer, Torenbeek) on the sizing loop and constraint analysis.

## How to read these notebooks critically
